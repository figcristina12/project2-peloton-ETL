```python
import pandas as pd
from pathlib import Path
```

```python
peloton_workouts=Path ('workouts.csv')
```

```python
peloton_workouts_df= pd.read_csv(peloton_workouts,low_memory=False)

#removed rows with all NaN values
peloton_workouts_df= peloton_workouts_df.dropna(how='all')

peloton_workouts_df.head (40000)
```

```python
peloton_workouts_df.info()
```

```python
peloton_workouts_df.drop(columns=['id','availability',
'class_type_ids',
'content_provider',
'content_format',
'difficulty_estimate',
'overall_estimate',
'difficulty_rating_avg',
'difficulty_rating_count',
'duration',
'equipment_ids',
'equipment_tags',
'explicit_rating',
'extra_images',
'fitness_discipline_display_name',
'has_closed_captions',
'has_pedaling_metrics',
'home_peloton_id',
'image_url',
'individual_instructor_ids',
'is_archived',
'is_closed_caption_shown', 
'has_free_mode',
'is_live_in_studio_only',
'language',
'origin_locale',
'length',
'live_stream_id', 
'live_stream_url',
'location',
'original_air_time',
'overall_rating_avg',
'overall_rating_count',
'pedaling_start_offset',
'pedaling_end_offset',
'pedaling_duration',
'rating', 
'ride_type_id',  
'ride_type_ids',
'sample_vod_stream_url',
'sample_preview_stream_url',
'scheduled_start_time',
'series_id', 
'sold_out',
'studio_peloton_id',
'total_ratings',
'total_in_progress_workouts',
'total_workouts', 
'vod_stream_url',
'vod_stream_id',
'captions',
'join_tokens',
'flags',
'is_dynamic_video_eligible',
'is_fixed_distance',
'dynamic_video_recorded_speed_in_mph',
'thumbnail_title', 
'thumbnail_location',
'distance',
'distance_unit',
'distance_display_value',
'muscle_group_score',
'is_sessions_eligible',
'total_user_workouts',
'total_following_workouts',
'is_favorite'], inplace=True)
```

```python
def encode_decode_text(text):
    if isinstance(text, str):

        return text.encode('latin-1', errors='ignore').decode('utf-8', errors='ignore')
    else:
        return text


peloton_workouts_df['description'] = peloton_workouts_df['description'].apply(encode_decode_text)
```

```python
peloton_workouts_df[['duration', 'workout_type']] = peloton_workouts_df['title'].str.extract(r'(\d+\s*min)\s*(.*)')
peloton_workouts_df['duration'] = peloton_workouts_df['duration'].str.strip()
peloton_workouts_df['workout_type'] = peloton_workouts_df['workout_type'].str.strip()


print(peloton_workouts_df[['duration', 'workout_type']])
```

```python
peloton_workouts_df.drop(columns=['title'
], inplace=True)
```

```python
peloton_workouts_df.rename(columns={'description': 'Class Description'}, inplace=True)

peloton_workouts_df.rename(columns={'difficulty_level': 'Level of Difficulty'}, inplace=True)

peloton_workouts_df.rename(columns={'fitness_discipline': 'Type of Exercise'}, inplace=True)

peloton_workouts_df.rename(columns={'is_explicit': 'Music w/ Explicit Lyrics'}, inplace=True)

peloton_workouts_df.rename(columns={'metrics': 'Health Impact'}, inplace=True)

peloton_workouts_df.rename(columns={'duration': 'Class Length'}, inplace=True)

peloton_workouts_df.rename(columns={'workout_type': 'Class Name'}, inplace=True)
```

```python
peloton_workouts_df.head (40000)
```

```python
peloton_workouts_df.info()
```

```python
# 1. Join the 2 datasets based on "id" and "instructor_id"
merged_df = peloton_workouts_df.merge(peloton_instructors_df, left_on='instructor_id', right_on='id', how='inner')

```

```python
merged_df['Instructor Name'] = merged_df['id'] + ' ' + merged_df['instructor_id']
```

```python
merged_df.sort_values(by='Instructor Name', inplace=True)
```

```python
merged_df.drop(columns=['id', 'instructor_id'], inplace=True)
```

```python
merged_df.rename(columns={'Class Description': 'Class_Description', 'Level of Difficulty': 'Level_of_Difficulty', 'Type of Exercise': 'Type_of_Exercise', 'Music w/ Explicit Lyrics': 'Music_w_Explicit_Lyrics', 'Class Length': 'Class_Length', 'Health Impact': 'Health_Impact', 'Class Name': 'Class_Name'}, inplace=True)
```

```python
merged_df.info()
```

```python
merged_df['Health Impact'] = merged_df['Health Impact'].str.replace('[', '', regex=False).str.replace(']', '', regex=False).str.replace("'", '', regex=False).str.replace('_', ' ', regex=True)

```

```python
merged_df.head (40000)
```

```python
merged_df.drop(columns=['Instructor Name'], inplace=True)
```

```python
merged_df.head (40000)
```

```python
merged_df.rename(columns={'name': 'Instructor Name'}, inplace=True)
```

```python
merged_df.head (40000)
```

```python
merged_df.rename(columns={
    'Class_Description': 'Class Description',
    'Level_of_Difficulty': 'Level of Difficulty',
    'Type_of_Exercise': 'Type of Exercise',
    'Music_w_Explicit_Lyrics': 'Music w/ Explicit Lyrics',
    'Health_Impact': 'Health Impact',
    'Class_Length': 'Class Length',
    'Class_Name': 'Class Name',
    'quote': 'Favorite Quote'
}, inplace=True)
```

```python
desired_columns = [
    'Instructor Name',
    'Favorite Quote',
    'Type of Exercise',
    'Level of Difficulty',
    'Music w/ Explicit Lyrics',
    'Class Length',
    'Health Impact',
    'Class Name',
    'Class Description'
]
```

```python
merged_df = merged_df[desired_columns]
```
